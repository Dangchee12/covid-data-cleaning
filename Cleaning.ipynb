```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("20220123 Global Covid Data.csv")
df.head()
```

```python
df.info()
```

```python
data_required = ["location", "date", "new_cases_smoothed", "new_deaths_smoothed", "new_cases_smoothed_per_million", "new_deaths_smoothed_per_million", "icu_patients_per_million", "hosp_patients_per_million", "weekly_icu_admissions_per_million", "weekly_hosp_admissions_per_million", "new_tests_smoothed_per_thousand", "people_fully_vaccinated_per_hundred", "total_boosters_per_hundred", "new_vaccinations_smoothed_per_million", "stringency_index", "population"]
df = df[data_required]
df.head()
```

```python
countries = ["United Kingdom", "Denmark", "Singapore", "Malaysia", "United States", "South Africa"]
df = df[df.location.isin(countries)]
df.head()
```

```python
df.info()
```

```python
df_uk = df[(df.date > "2021-04-30") & (df.location == "United Kingdom")]
df_malaysia = df[(df.date > "2021-04-30") & (df.location == "Malaysia")]
df_singapore = df[(df.date > "2021-04-30") & (df.location == "Singapore")]
df_denmark = df[(df.date > "2021-04-30") & (df.location == "Denmark")]
df_southafrica = df[(df.date > "2021-04-30") & (df.location == "South Africa")]
df_us = df[(df.date > "2021-04-30") & (df.location == "United States")]
```

```python
frames = [df_uk, df_malaysia, df_singapore, df_denmark, df_southafrica, df_us]
country = ["uk", "malaysia", "singapore", "denmark", "southafrica", "us"]
i = 0
for frame in frames:
    frame.to_csv(country[i] + ".csv")
    i+=1
    
```

```python
from datetime import date, timedelta

df_uk = df_uk.set_index("date")

sdate = date(2021,9,1)
edate = date(2021,11,30)
period = pd.date_range(sdate,edate,freq='d').strftime('%Y-%m-%d').tolist()
df_uk.drop(labels=period, axis=0, inplace=True)

df_uk["wave"] = "delta"
sdate = date(2021,12,1)
edate = date.today() - timedelta(days=1)

period = pd.date_range(sdate,edate,freq='d').strftime('%Y-%m-%d').tolist()
df_uk.at[period, "wave"] = "omicron"

df_uk.reset_index(inplace=True)
df_uk.head()
```
